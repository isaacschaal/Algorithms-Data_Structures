```python
import random


def hat_party(n):
    new_hat_list =[]
    real_hat_list = range(0,n)
    hats = range(0,n)
    for x in real_hat_list:
        a = random.choice(hats)
        new_hat_list.append(a)
        y = hats.index(new_hat_list[x])
        del(hats[y])
    return new_hat_list

print hat_party(5)
```

```python
xlis = range(1,1000,10)
ylis = []
repeats = 100

for n in xlis:
    avg = []
    for _ in range(repeats):
        new_hat_list = hat_party(n)
        hat_counter = 0
        real_hat_list = range(0,n)
        for x in range(len(new_hat_list)):
            if new_hat_list[x] == real_hat_list[x]:
                hat_counter += 1
        avg.append(hat_counter)   
    ylis.append(np.mean(avg))
```
