```python
import heapq
import numpy as np
import math


# min heap on top, max heap on other side

def median(minh, maxh):
    """The heaps must come in at same length, or one 
    or the other one greater than the other"""
    if len(minh) > len(maxh):
        return minh[0]
        
    elif len(maxh) > len(minh):
        return -1*maxh[0]
    else:
        return np.mean([minh[0],-1*maxh[0]])

def add_to_median_heap(minh,maxh,elem):
    if len(minh) == 0:
        heapq.heappush(minh,elem)
    else:
        med = median(minh,maxh)
        if med <= elem:
            heapq.heappush(minh,elem)
        else:
            heapq.heappush(maxh,-1*elem)
        if len(minh) == len(maxh):
            pass
        else:
            if len(minh) - len(maxh) > 1:
                heapq.heappush(maxh,-1*heapq.heappop(minh))
            elif len(maxh) - len(minh) > 1:
                heapq.heappush(minh,-1*heapq.heappop(maxh))
minh = []
maxh = []


for a in range(1,100,2):
    add_to_median_heap(minh, maxh, a)
    #print(median(minh, maxh))
    
## It is possible to do any percentile, we must just adjust 
# the 
```
