```python
#Learn how to record the time a script takes to run

import time
# start_time = time.time()
# print "My program took", time.time() - start_time, "to run"

#understand insertion sort, selection sort and bubble sort

"""Insertion sort -- takes the list, takes an element, searchs to the left
of it to find the place that it goes in, and then moves to the next number
*searches through the sorted list to find the place for the number, choosing
the number just to the right of the list*"""


"""selection sort -- find the smallest element, put it in place 1,
find the second smallest, put it in place 2, so on"""

"""Bubble sort --- starts with the right most number, and switches it
with the one to the left it its smaller, then does that with one
place to the left, all the way to i.. then it increases i by 1, does the 
whole thing over... so if its the smallest number, it will move all 
the way to the beggining, it just keeps switching them"""

#implement INSERTION SORT and one of the others such that
#sorts list that contain integers
#prints its own running time and a count of number of steps it makes

"""am still confused about how to do the 'number of steps'
that is required for these functions, at the end of each loop?
or each line?"""


test = [5,1,3,4,2,6,11,9]
test2 = [5,1,3,4,2,6,11,9]



def insertion_sort(lis):
    start_time = time.time()
    steps = 0
    #runs through all elements in list, starting with the second
    for j in range(1,len(lis)):
        #sets the key, the number being arranged
        key = lis[j]
        #sets i, the number that the key will be compared to
        i= j-1
        # if i is still in the list, and the number at i
        # is greater than the key
        while i>=0 and lis[i]>key:
            # move i one position to the right
            lis[i+1] = lis[i]
            # change the number for the key to be compared to one left
            i += -1
            
            # the while loop then repeats the process, untill either
            # the key is the smallest number (i goes to -1) or the key
            # is greater than (or equal to) a number and we have found
            # its correct place in the list
            steps +=1
        # places the key in its correct position (i+1 because its
        # one to the right of the first number it is bigger than)
        lis[i+1]=key 
        steps += 1
    print "My program took", time.time() - start_time, "to run"
    print "in", steps, "steps"
    return lis

def bubble_sort(lis):
    start_time = time.time()
    steps = 0
    #iterates through all elements, starting with the first untill the
    #second to last, each time this loop moves, one more on the left is sorted
    for i in (range(len(lis)-1)):
        # iterates from the fartherst right until the edge of the sorted list
        # switches any number that needs to be switched
        # and ends up moving the smallest number to the left
        # and creating a sorted list on the left
        for j in range(len(lis)-1,i,-1):
            if (lis[j] < lis[j-1]):
                temp = lis[j]
                lis[j] = lis[j-1]
                lis[j-1] = temp
            steps += 1
        steps += 1
    print "My program took", time.time() - start_time, "to run"
    print "in", steps, "steps"
    return lis


#Learn worst case scenario of selection and bubble sort in big O

"""for bubble, is N^2, also for selection is N^2"""


#know the worst case running time for insertion sort
"""insertion is N^2 as well"""


#write order of growth for T(n)=2n^4+3(n^2)/5-n+1 in big O,
#is it faster or slower than insertion sort? in All cases?

"""n^4, should be slower, no, in vary small cases it could be faster?"""

print insertion_sort(test)
print bubble_sort(test2)


```
