```python
import timeit
import random

eps = 1e-16
N = 10000
locations = [0.0, 0.5, 1.0 - eps]


def median(x1, x2, x3):
    for a in range(7):
        if x1 <= x2 <= x3:
            return x2
        # Every loop I'm shufflin
        (x1, x2, x3) = (x2, x1, x3)
        if a % 2:
            (x1, x2, x3) = (x3, x1, x2)


def qsort(lst):
    indices = [(0, len(lst))]

    while indices:
        (frm, to) = indices.pop()
        if frm == to:
            continue

        # Find the partition:
        N = to - frm
        inds = [frm + int(N * n) for n in locations]
        values = [lst[ind] for ind in inds]
        partition = median(*values)

        # Split into lists:
        lower = [a for a in lst[frm:to] if a < partition]
        upper = [a for a in lst[frm:to] if a > partition]
        counts = sum([1 for a in lst[frm:to] if a == partition])

        ind1 = frm + len(lower)
        ind2 = ind1 + counts

        # Push back into correct place:
        lst[frm:ind1] = lower
        lst[ind1:ind2] = [partition] * counts
        lst[ind2:to] = upper

        # Enqueue other locations
        indices.append((frm, ind1))
        indices.append((ind2, to))
    return lst


def randomized_quicksort():
    lst = list(range(N))
    random.shuffle(lst)
    return qsort(lst)


def test_quicksort():
    lst = randomized_quicksort()
    assert (lst == list(range(N)))


# Is our algorithm correct
test_quicksort()

# How fast is our algorithm
print (timeit.timeit(randomized_quicksort, number=1))
```

```python
import timeit
import random

eps = 1e-16
N = 1000000
locations = [0.0, 0.5, 1.0 - eps]


def median(x1, x2, x3):
    for a in range(7):
        if x1 <= x2 <= x3:
            return x2
        # Every loop I'm shufflin
        (x1, x2, x3) = (x2, x1, x3)
        if a % 2:
            (x1, x2, x3) = (x3, x1, x2)

def partition(lst,frm,to):
    # Find the partition:
    N = to - frm
    inds = [frm + int(N * n) for n in locations]
    values = [lst[ind] for ind in inds]
    partition = median(*values)

    # Split into lists:
    lower = [a for a in lst[frm:to] if a < partition]
    upper = [a for a in lst[frm:to] if a > partition]
    counts = sum([1 for a in lst[frm:to] if a == partition])

    ind1 = frm + len(lower)
    ind2 = ind1 + counts

    # Push back into correct place:
    lst[frm:ind1] = lower
    lst[ind1:ind2] = [partition] * counts
    lst[ind2:to] = upper
    return ind1,ind2

def qsort(lst,frm,to):
    if frm<to:
        ind1,ind2 = partition(lst,frm,to)
        qsort(lst,frm,ind1)
        qsort(lst,ind2,to)
    return lst

def randomized_quicksort():
    lst = list(range(N))
    random.shuffle(lst)
    return qsort(lst,0,N)


def test_quicksort():
    lst = randomized_quicksort()
    assert (lst == list(range(N)))


# Is our algorithm correct
test_quicksort()

# How fast is our algorithm
print (timeit.timeit(randomized_quicksort, number=1))

```
